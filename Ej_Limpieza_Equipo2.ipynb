## Tarea extra
#### df.dropna(): La funci√≥n dropna nos permite eliminar las filas o columnas que contienen algun valor #NA, se tienen 3 formas: 
- df.dropna() la cual elimina la fila donde se tiene el valor. 
- df.dropna(axis=1) es la que elimina la columna donde se encuentra el #NA.
- df.dropna(how="all") elimina las filas donde todos los valores sean #NA.

#### df.astype(): Se utiliza para cambiar el tipo de datos de una columna entera se utiliza df.name=df.name.astype('type'), donde name es el nombre de la columna y type el tipo de dato al que queremos cambiarlo.
