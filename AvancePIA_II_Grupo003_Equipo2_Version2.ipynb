# 2) Descripción de los datos


a) Tipo de datos.

La base de datos esta constituida por indicadores demográficos, las columnas estan conformadas por tipo de dato númerico excepto la columna de país y de estatus ya que son de tipo texto. 


b) Descripción de las columnas

**País:** Lugar del cual se tomaron los datos.

**Anio:** Indica el año en que se recopilaron los datos, por país se tomaron en cuenta del año 2000 al 2015. 

**Estatus:** Indica si el país es desarrollado (tiene un alto nivel de vida) o está en desarrollo (economía en transición).

**Esperanza_de_Vida:** Esperanza de vida promedio al nacer, en años.

**Mortalidad de adultos:** Número de muertes de adultos de ambos sexos entre 15 y 60 años por cada 1000 habitantes.

**Mortalidad_infantil:** Número de muertes de niños por cada 1000 nacimientos vivos registrados, durante el primer año de vida.

**Alcohol:** Registro de consumo de alcohol medido en litros de alcohol puro per cápita por año.

**Porcentaje_de_gastos:** Gasto en salud como porcentaje del Producto Interno Bruto per cápita.

**Hepatitis_B:** Cobertura de inmunización por hepatitis B (HepB) en porcentaje de niños de 1 año.

**Sarampion:** Número de casos de sarampión notificados por cada 1000 habitantes.

**Indice_Masa_Corporal:** Índice promedio de masa corporal de toda la población.

**Muertes_niños_menores_a_5anios:** Número de muertes de niños menores de cinco años por cada 1000 habitantes.

**Poliomelitis:** Cobertura de inmunización contra la poliomielitis (Pol3) en porcentaje de niños de 1 año. 

**Gasto_total:** Gasto público general en salud como porcentaje del gasto público total.

**Difteria:** Cobertura de inmunización por difteria tétanos y tos ferina (DTP3) en porcentaje de niños de 1 año.
	
**VIH/SIDA:** Muertes por cada 1 000 nacidos vivos VIH/SIDA (0-4 años).

**PIB:** Producto Interno Bruto per cápita (en USD).

**Población:** Población total del país.

**Delgadez_1a19_anios:** Prevalencia de delgadez entre niños y adolescentes de 10 a 19 años en porcentaje

**Delgadez_5a9_anios:** Prevalencia de delgadez entre niños de 5 a 9 años en porcentaje.

**Composicion_de_los_ingresos_de_los_recursos:** Índice de Desarrollo Humano en términos de composición de ingresos de los recursos (índice que va de 0 a 1).

**Escolarización:** Número de años de escolarización.
